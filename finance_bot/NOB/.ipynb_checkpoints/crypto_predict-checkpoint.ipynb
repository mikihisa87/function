```python
import psycopg2
import subprocess
import os
import glob
import pandas as pd
import time
from enum import Enum

def get_conn_url():
    url = 'postgresql://{}:{}@{}:5432/{}'.format(
            'crypto',
            'analyzer',
            '192.168.0.30',
            'crypto_analyzer_dev')
    return url

def conn(n_try=5):
    exception = None
    while n_try > 0:
        try:
            return psycopg2.connect(get_conn_url())
        except Exception as e:
            exception = e
            time.sleep(0.5)
            n_try -= 1
    print(exception)

def do_sql(sql):
    c = conn()
    with c.cursor() as cursor:
        try:
            cursor.execute(sql)
            c.commit()
        except Exception as e:
            print(e)

def fetch_all_sql(sql):
    c = conn()
    with c.cursor() as cursor:
        try:
            cursor.execute(sql)
            result = cursor.fetchall()
            c.commit()
            return result
        except Exception as e:
            print(e)
            return None

def fetch_one_sql(sql):
    c = conn()
    with c.cursor() as cursor:
        try:
            cursor.execute(sql)
            result = cursor.fetchone()
            c.commit()
            return result
        except Exception as e:
            print(e)
            return None

def execute_sql_path(sql_path):
    from crypto_analyzer.config import app_config
    print('\nexecute {}...'.format(os.path.basename(sql_path)))
    db = app_config().DB
    args = db.connect_kwargs
    command = 'psql -h {} -U {} -p 5432 {} < {}'.format(
            args['host'],
            args['user'],
            db.database,
            sql_path)
    print(subprocess.getoutput(command))

def do_migration():
    dir_name = os.path.dirname(os.path.abspath(__file__))
    for sql_path in sorted(glob.glob(os.path.join(dir_name, 'migrations', '*.sql'))):
        execute_sql_path(sql_path)

class Query:
    class Type(Enum):
        INSERT = 'INSERT'
        SELECT_MANY = 'SELECT_MANY'
        SELECT_ONE = 'SELECT_ONE'
        PANDAS = 'PANDAS'

    def __init__(self):
        self.sql = None
        self.mode = 'normal'

    def query(self, query):
        self.sql = query
        return self

    def fetch_mode(self, mode):
        self.mode = mode
        return self

    def run(self):
        if self.sql is None:
            return None

        if self.mode is Query.Type.PANDAS:
            return pd.read_sql_query(self.sql, get_conn_url())
        elif self.mode is Query.Type.SELECT_MANY:
            return fetch_all_sql(self.sql)
        elif self.mode is Query.Type.SELECT_ONE:
            return fetch_one_sql(self.sql)
        elif self.mode is Query.Type.INSERT:
            return do_sql(self.sql)
        else:
            return None

def candlestick_prices(period, currency_code, from_time, to_time, offset, limit,
        desc=False, exchange=None):
    order_by = 'DESC' if desc else ''
    where_exchange = "AND exchange = '{}'".format(exchange) if exchange else ''
    sql = """
    SELECT time_bucket('{}', time) AS period,
    last(last, time) as v_t,
    MAX(last) as v_hi,
    MIN(last) as v_lo,
    last(bid, time) as bid,
    last(ask, time) as ask
    FROM exchange_tickers
    WHERE currency_code = '{}' AND '{}' <= time AND time <= '{}' {}
    GROUP BY period
    ORDER BY period {}
    OFFSET {}
    LIMIT {};
    """.format(period, currency_code, from_time, to_time, where_exchange, order_by, offset, limit)
    return Query().query(sql).fetch_mode(Query.Type.PANDAS).run()
```

```python
# Feature Engineering 移動平均
#5足
data2 = np.c_[data2, np.zeros((len(data2), 1))] #列の追加
ave_num = 5
for i in range(ave_num, len(data2)):
    tmp = data2[i - ave_num + 1: i + 1, 4].astype(np.float) # 4=v_t
    data2[i, 5] = np.mean(tmp) #挿入場所

#25足
data2 = np.c_[data2, np.zeros((len(data2), 1))]
ave_num = 25
for i in range(ave_num, len(data2)):
    tmp = data2[i - ave_num + 1: i + 1, 4].astype(np.float)
    data2[i, 6] = np.mean(tmp)

#75足
data2 = np.c_[data2, np.zeros((len(data2), 1))]
ave_num = 75
for i in range(ave_num, len(data2)):
    tmp = data2[i - ave_num + 1: i + 1, 4].astype(np.float)
    data2[i, 7] = np.mean(tmp)

#200足
data2 = np.c_[data2, np.zeros((len(data2), 1))]
ave_num = 200
for i in range(ave_num, len(data2)):
    tmp = data2[i - ave_num + 1: i + 1, 4].astype(np.float)
    data2[i, 8] = np.mean(tmp)
```

```python
# Feature Engineering 一目均衡表(9,26, 52) (2, 5, 8)
para1 =9
para2 = 26
para3 = 52

# 転換線 = （過去(para1)回分間の高値 + 安値） ÷ 2
data2 = np.c_[data2, np.zeros((len(data2),1))] # 列の追加
for i in range(para1, len(data2)):
    tmp_high =data2[i-para1+1:i+1,2].astype(np.float) # 2 = v_hi
    tmp_low =data2[i-para1+1:i+1,3].astype(np.float) # 3 = v_lo
    data2[i,9] = (np.max(tmp_high) + np.min(tmp_low)) / 2 
    
# 基準線 = （過去(para2)回分間の高値 + 安値） ÷ 2
data2 = np.c_[data2, np.zeros((len(data2),1))]
for i in range(para2, len(data2)):
    tmp_high =data2[i-para2+1:i+1,2].astype(np.float)
    tmp_low =data2[i-para2+1:i+1,3].astype(np.float)
    data2[i,10] = (np.max(tmp_high) + np.min(tmp_low)) / 2 

# 先行スパン1 = ｛ （転換値+基準値） ÷ 2 ｝を(para2)回数先にずらしたもの
data2 = np.c_[data2, np.zeros((len(data2),1))]
for i in range(0, len(data2)-para2):
    tmp =(data2[i,9] + data2[i,10]) / 2 
    data2[i+para2,11] = tmp


# 先行スパン2 = ｛ （過去(para3)回分間の高値+安値） ÷ 2 ｝を(para2)回分先にずらしたもの
data2 = np.c_[data2, np.zeros((len(data2),1))]
for i in range(para3, len(data2)-para2):
    tmp_high =data2[i-para3+1:i+1,2].astype(np.float)
    tmp_low =data2[i-para3+1:i+1,3].astype(np.float)
    data2[i+para2,12] = (np.max(tmp_high) + np.min(tmp_low)) / 2
```

```python
# Feature Engineering 25回分ボリンジャーバンド（±1, 2シグマ）
parab = 25
data2 = np.c_[data2, np.zeros((len(data2),4))] # 列の追加
for i in range(parab, len(data2)):
    tmp = data2[i-parab+1:i+1,4].astype(np.float)
    data2[i,13] = np.mean(tmp) + 1.0* np.std(tmp) 
    data2[i,14] = np.mean(tmp) - 1.0* np.std(tmp) 
    data2[i,15] = np.mean(tmp) + 2.0* np.std(tmp) 
    data2[i,16] = np.mean(tmp) - 2.0* np.std(tmp)
```

```python
# 【重要】X, Yを正規化
original_X = np.copy(X) # コピー
tmp_mean = np.zeros(len(X))

for i in range(num_ago,len(X)):
    tmp_mean[i] = np.mean(original_X[i-num_ago+1:i+1,0]) # 回数分の平均値
    for j in range(0, X.shape[1]): 
        X[i,j] = (X[i,j] - tmp_mean[i]) # Xを正規化
    Y[i] =  Y[i]  # X同士の引き算しているので、Yはそのまま
```

```python
#予測結果の合計を計算
# 前々回終値に比べて前回終値が高い場合は、買いとする

#SVR
total_svr = 0
for i in range(0,len(Y_test_)): # len()で要素数を取得
    if Y_pred_svr[i] >= 0:
        total_svr += Y_test_[i]
    else:
        total_svr -= Y_test_[i]
print("SVR利益合計：%1.3lf" %total_svr) 

# 予測結果の総和グラフ
total_return_svr = np.zeros(len(Y_test_))
if Y_pred_svr[i] >=0: #初回を格納
    total_return_svr[0] = Y_test_[i]
else:
    total_return_svr[0] = -Y_test_[i]

for i in range(1, len(result_svr)): # 2回目以降を格納
    if Y_pred_svr[i] >=0:
        total_return_svr[i] = total_return_svr[i-1] + Y_test_[i]
    else:
        total_return_svr[i] = total_return_svr[i-1] - Y_test_[i]

#Linear
total = 0
for i in range(0,len(Y_test_)): # len()で要素数を取得
    if Y_pred[i] >= 0:
        total += Y_test_[i]
    else:
        total -= Y_test_[i]
print("Linear利益合計：%1.3lf" %total) 

# 予測結果の総和グラフ
total_return = np.zeros(len(Y_test_))
if Y_pred[i] >=0: #初回を格納
    total_return[0] = Y_test_[i]
else:
    total_return[0] = -Y_test_[i]

for i in range(1, len(result)): # 2回目以降を格納
    if Y_pred[i] >=0:
        total_return[i] = total_return[i-1] + Y_test_[i] 
    else:
        total_return[i] = total_return[i-1] - Y_test_[i]

#Ridge
total_ridge = 0
for i in range(0,len(Y_test_)): # len()で要素数を取得
    if Y_pred_ridge[i] >= 0:
        total_ridge += Y_test_[i]
    else:
        total_ridge -= Y_test_[i]
print("Ridge利益合計：%1.3lf" %total_ridge) 

# 予測結果の総和グラフ
total_return_ridge = np.zeros(len(Y_test_))
if Y_pred_ridge[i] >=0: #初回を格納
    total_return_ridge[0] = Y_test_[i]
else:
    total_return_ridge[0] = -Y_test_[i]

for i in range(1, len(result_ridge)): # 2回目以降を格納
    if Y_pred_ridge[i] >=0:
        total_return_ridge[i] = total_return_ridge[i-1] + Y_test_[i]
    else:
        total_return_ridge[i] = total_return_ridge[i-1] - Y_test_[i]

#Lasso
total_lasso = 0
for i in range(0,len(Y_test_)): # len()で要素数を取得
    if Y_pred_lasso[i] >= 0:
        total_lasso += Y_test_[i]
    else:
        total_lasso -= Y_test_[i]
print("Lasso利益合計：%1.3lf" %total_lasso) 

# 予測結果の総和グラフ
total_return_lasso = np.zeros(len(Y_test_))
if Y_pred_lasso[i] >=0: #初回を格納
    total_return_lasso[0] = Y_test_[i]
else:
    total_return_lasso[0] = -Y_test_[i]

for i in range(1, len(result_lasso)): # 2回目以降を格納
    if Y_pred_lasso[i] >=0:
        total_return_lasso[i] = total_return_lasso[i-1] + Y_test_[i]
    else:
        total_return_lasso[i] = total_return_lasso[i-1] - Y_test_[i]
        
plt.figure(figsize=(8,4.1))
plt.xlabel("Times")
plt.ylabel("Profit_Loss")
plt.plot(total_return_svr, label = "SVR")
plt.plot(total_return, label = "Linear")
plt.plot(total_return_ridge, label="Ridge")
plt.plot(total_return_lasso, label="Lasso")
plt.legend()
plt.show()
```
